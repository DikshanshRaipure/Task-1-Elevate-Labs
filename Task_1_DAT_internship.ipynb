```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files
uploaded = files.upload()

netflix = pd.read_csv('netflix_titles.csv')
```

```python
print("Netflix Head")
print(netflix.head())
print()

#data types and missing values
print("Netflix Info")
print(netflix.info())
print()

#count missing values in column
print("Netflix Missing Values")
print(netflix.isnull().sum())
print()

print(len(netflix))
```

```python
# Fill missing 'director' and 'cast' with 'No Data'
netflix['director'] = netflix['director'].fillna('No Data')
netflix['cast'] = netflix['cast'].fillna('No Data')

# Fill missing 'country' with the most common country (mode)
most_common_country = netflix['country'].mode()[0]
netflix['country'] = netflix['country'].fillna(most_common_country)

# Drop rows with missing 'date_added' or 'rating'
netflix = netflix.dropna(subset=['date_added', 'rating'])
```

```python
#removing duplicate rows, if any
netflix.drop_duplicates(inplace=True)
```

```python
# Standardize column names
netflix.columns = netflix.columns.str.lower().str.replace(' ', '_')

# Convert 'date_added' to datetime format
# Strip leading/trailing whitespace before conversion and use errors='coerce'
netflix['date_added'] = pd.to_datetime(netflix['date_added'].str.strip(), errors='coerce')
#the earlier code I tried had an error where one of the data was not in the correct format for Pnada to detect it and covert
#to standart date and time format. and so google colab suggested this change in code.

# Ensure 'release_year' is integer
netflix['release_year'] = netflix['release_year'].astype(int)

# (Optional) Check and standardize 'type' and 'rating' capitalization
netflix['type'] = netflix['type'].str.title()
netflix['rating'] = netflix['rating'].str.upper()
```

```python
#confirming the changes made
print(netflix.isnull().sum())

print(netflix.dtypes)

print(netflix.head())

print(len(netflix))
```
